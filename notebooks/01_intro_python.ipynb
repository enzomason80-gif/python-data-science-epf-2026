# Python for Data Science

## Notebook 01 — Mise en place de l'environnement de travail et introduction à Python

Objectifs de ce notebook :

- Comprendre l'environnement de travail utilisé dans ce cours
- Rappeler les outils utilisés : Python, Jupyter, Git et GitHub
- Comprendre les commandes Git essentielles
- Comprendre la syntaxe Python
- Manipuler les structures de données simples

## Environnement de travail du cours

Dans ce cours, nous utilisons plusieurs outils pour programmer et analyser des données :

- Python
- Jupyter Notebook
- Git
- GitHub

Ces outils permettent :

- d'écrire et exécuter du code Python
- d'analyser des données
- de versionner votre travail
- de sauvegarder votre travail en ligne

## Python

Python est le langage utilisé dans ce cours pour l'analyse de données.

Vous avez installé Python sur votre ordinateur.

Pour vérifier que Python est correctement installé, ouvrez un terminal et tapez :

python --version

Si Python est installé correctement, vous devriez voir une version de Python, par exemple :

Python 3.11.2

## Jupyter Notebook

Jupyter Notebook est un environnement interactif qui permet :

- d'écrire du code Python
- d'exécuter ce code cellule par cellule
- d'ajouter du texte et des explications
- d'afficher des résultats et des graphiques

Pour lancer Jupyter Notebook depuis votre terminal :

jupyter notebook

Un navigateur web s'ouvre automatiquement et vous permet d'accéder aux notebooks.

## GitHub

GitHub est une plateforme qui permet d'héberger des dépôts Git en ligne.

Dans ce cours, GitHub sera utilisé pour :

- sauvegarder votre travail
- conserver l'historique de vos modifications
- partager vos projets

## Git

Git est un système de gestion de versions.

Il permet de :

- sauvegarder différentes versions de votre travail
- suivre les modifications d'un projet
- collaborer sur des projets

Pour vérifier que Git est installé sur votre ordinateur :

git --version

## Repository du cours

Le repository du cours contient :

- les notebooks de cours
- les datasets utilisés
- les exercices

Vous avez créé une copie de ce repository dans votre compte GitHub en utilisant la fonctionnalité **Fork**.

Le fork permet de créer votre propre version du repository du cours.

## Cloner votre repository

Après avoir créé votre fork, vous avez cloné votre repository sur votre ordinateur.

La commande utilisée est :

git clone https://github.com/votre-username/python-data-science-epf-2026.git

Cette commande crée une copie locale du repository sur votre machine.

Ensuite vous entrez dans le dossier du projet :

cd python-data-science-epf-2026

## Organisation du projet

Le repository est organisé de la manière suivante :

datasets/
notebooks/
exercises/
project/

- datasets contient les données utilisées dans le cours
- notebooks contient les notebooks de cours
- exercises contient les exercices
- project sera utilisé pour un projet final (TP)

### Début de séance

Au début de chaque séance, vous pouvez récupérer les mises à jour du repository du cours :

git pull https://github.com/Pierrotlefoot/python-data-science-epf-2026.git main

Cela permet de récupérer :

- les nouveaux notebooks
- les datasets
- les corrections éventuelles

### Fin de séance

À la fin de chaque séance, vous pouvez sauvegarder votre travail sur GitHub :

- git add .
- git commit -m "Travail séance X"
- git push

Ces commandes permettent :

- d'ajouter les fichiers modifiés
- de créer une nouvelle version de votre travail
- d'envoyer cette version sur votre repository GitHub

## Authentification GitHub

Dans ce cours, nous utilisons Git via l'URL HTTPS.

Lorsque vous utilisez la commande :

git push

GitHub peut vous demander de vous authentifier.

Dans ce cas, une fenêtre de connexion GitHub peut s'ouvrir dans votre navigateur.

Il suffit de vous connecter à votre compte GitHub.

## Commandes Git essentielles

Voici les principales commandes Git utilisées dans ce cours.

| Commande | Utilité |
|--------|--------|
| git clone | Copier un repository GitHub sur votre ordinateur |
| git pull | Récupérer les mises à jour d'un repository |
| git add | Ajouter des fichiers modifiés à la prochaine version |
| git commit | Enregistrer une nouvelle version de votre travail |
| git push | Envoyer votre travail sur GitHub |
| git status | Voir les fichiers modifiés |
| git log | Voir l'historique des versions |

## Vérifier l'état du repository

Pour voir les fichiers modifiés dans votre repository :

git status
