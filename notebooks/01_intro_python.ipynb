## Structures de données Python

Python propose plusieurs structures de données pour stocker et organiser l'information.

Dans ce cours, nous allons manipuler :

- les listes
- les tuples
- les dictionnaires
- les chaînes de caractères
- les ensembles

Chaque structure a des propriétés différentes.

### 1. Les listes

Une liste permet de stocker plusieurs éléments dans un ordre précis.

Une liste est **modifiable**.

```python
numbers = [1, 2, 3, 4, 5]

print(numbers)
print(type(numbers))
```

```python
print(numbers[0])
print(numbers[2])
```

```python
numbers[0] = 10
print(numbers)
```

Une liste :

- conserve l'ordre des éléments
- peut contenir des doublons
- peut être modifiée

### 2. Les tuples

Un tuple ressemble à une liste, mais il est **immutable** : on ne peut pas le modifier après sa création.

```python
coordinates = (10, 20)

print(coordinates)
print(type(coordinates))
```

```python
print(coordinates[0])
```

```python
coordinates[0] = 5
```

Cette dernière instruction provoque une erreur, car un tuple ne peut pas être modifié.

Un tuple :

- conserve l'ordre des éléments
- peut contenir des doublons
- ne peut pas être modifié

### 3. Les dictionnaires

Un dictionnaire stocke des données sous forme de **clé → valeur**.

```python
student = {
    "name": "Alice",
    "age": 21,
    "grade": 15
}

print(student)
print(type(student))
```

```python
print(student["name"])
print(student["age"])
```

```python
student["age"] = 22
print(student)
```

```python
student["city"] = "Paris"
print(student)
```

Un dictionnaire :

- associe une clé à une valeur
- ne s'utilise pas avec des positions comme une liste
- est modifiable
- utilise des clés uniques

### 4. Les chaînes de caractères

Une chaîne de caractères représente du texte.

```python
name = "Alice"

print(name)
print(type(name))
```

```python
print(name[0])
print(name[1])
```

```python
print(len(name))
print(name.upper())
print(name.lower())
```

Une chaîne :

- représente du texte
- conserve l'ordre des caractères
- n'est pas modifiable directement

```python
name[0] = "a"
```

Cette instruction provoque une erreur, car une chaîne de caractères est immutable.

### 5. Les ensembles

Un ensemble est une collection d'éléments **sans ordre** et **sans doublons**.

```python
values = {1, 2, 3, 3, 4, 4, 5}

print(values)
print(type(values))
```

Les doublons disparaissent automatiquement.

```python
values.add(6)
print(values)
```

Un ensemble :

- ne conserve pas l'ordre
- ne contient pas de doublons
- est utile pour éliminer les répétitions

## Comparaison des structures

| Structure | Ordre | Modifiable | Doublons |
|----------|-------|------------|----------|
| Liste | oui | oui | oui |
| Tuple | oui | non | oui |
| Dictionnaire | clés | oui | clés uniques |
| String | oui | non | oui |
| Set | non | oui | non |

```python
numbers = [1, 2, 3]
coordinates = (10, 20)
student = {"name": "Alice", "age": 21}
firstname = "Alice"
values = {1, 2, 2, 3}

print("Liste :", numbers)
print("Tuple :", coordinates)
print("Dictionnaire :", student)
print("Chaîne :", firstname)
print("Set :", values)
```
