# 1. NumPy : tableaux numériques

## Pourquoi NumPy ?

Une liste Python est pratique, mais elle n'est pas idéale pour le calcul numérique.

**NumPy** permet de :
- stocker des nombres dans des tableaux efficaces
- faire des calculs rapidement
- appliquer des opérations sur tout un tableau d'un seul coup

C'est une bibliothèque de base en data science.


Commençons par la base, importer la librairie NumPy :

Lorsqu'on fait cette opération, Python va chercher la bibliothèque (ou le module) NumPy dans notre environnement ou notre système et exécute son code interne : notamment, en faisant ça on (l'ordinateur) définit les fonctions associées à NumPy, ce qui va nous permettre de les appeler plus tard.

On les appellera en utilisant le diminutif qu'on a donné à NumPy, np (diminutif classique pour éviter la lourdeur d'écriture). De la même manière qu'on appelait par exemple la méthode "append" des listes en écrivant maliste.append(), ici on fera np.la_méthode(les_arguments). La seule différence est qu'avec NumPy, on appelle des fonctions liées à un module, et non pas des méthodes liées à un objet (les listes sont un objet) : c'est une différence de langage.

```python
import numpy as np
```

## Différence entre liste Python et tableau NumPy

Une différence importante :

Avec une **liste Python**, l'opération `* 2` répète la liste.  
Avec un **tableau NumPy**, l'opération `* 2` multiplie chaque valeur par 2.

C'est le principe de la **vectorisation**.

```python
liste = [1, 2, 3]
tableau = np.array([1, 2, 3])

print("Liste Python * 2 :", liste * 2)
print("Tableau NumPy * 2 :", tableau * 2)
```

## Indexation et slicing

Comme avec les listes, on peut accéder à certains éléments :
- `arr[0]` : premier élément

```python
arr = np.array([10, 20, 30, 40, 50])

print("Premier élément :", arr[0])
print("Dernier élément :", arr[-1])
print("Slice [1:4] :", arr[1:4])
```

## Filtrage conditionnel

```python
arr = np.array([10, 20, 30, 40, 50])

print("Tableau complet :", arr)
print("Valeurs > 25 :", arr[arr > 25])
```

# 3. Matplotlib : visualiser des données

## Pourquoi visualiser ?

Les tableaux donnent des informations utiles, mais les graphiques permettent souvent de :
- voir une distribution
- repérer des tendances
- comparer des groupes
- mieux interpréter les résultats

Pour cela, on utilise la bibliothèque Matplotlib (et Seaborn en plus avancé, que nous verrons par la suite), et nous pouvons nous appuyer sur ce tutoriel : https://www.geeksforgeeks.org/python/matplotlib-tutorial/

## Rappel

Avec Matplotlib, on suit souvent ce schéma :

1. créer une figure ;
2. tracer un graphique ;
3. ajouter un titre et des labels ;
4. afficher avec `plt.show()`.


## Histogramme : distribution des âges


```python
plt.figure(figsize=(8, 5))
plt.hist(df["Age"].dropna(), bins=20, color = 'g')

plt.title("Distribution des âges")
plt.xlabel("Âge")
plt.ylabel("Effectif")
plt.grid(True)

plt.show()
```

## Diagramme en barres : taux de survie par sexe


```python
survival_by_sex = df.groupby("Sex")["Survived"].mean()

plt.figure(figsize=(8, 5))
plt.bar(survival_by_sex.index, survival_by_sex.values)

plt.title("Taux de survie par sexe")
plt.xlabel("Sexe")
plt.ylabel("Survie moyenne")
plt.grid(axis="y")

plt.show()
```

## Diagramme en barres avec Pandas

Pandas peut s'appuyer sur Matplotlib pour tracer rapidement un graphique.


```python
df.groupby("Pclass")["Survived"].mean().plot(kind="bar", figsize=(8, 5))

plt.title("Taux de survie par classe")
plt.xlabel("Classe")
plt.ylabel("Survie moyenne")
plt.grid(axis="y")

plt.show()
```

## Scatter plot : âge et prix du billet

Un **scatter plot** sert à observer la relation entre deux variables numériques.


```python
plt.figure(figsize=(8, 5))
plt.scatter(df["Age"], df["Fare"], alpha=0.5)

plt.title("Âge et prix du billet")
plt.xlabel("Âge")
plt.ylabel("Prix du billet")
plt.show()
```

## Exercice Matplotlib 1 - Comparaison croisée

Comparer le taux de survie selon :
- le sexe
- et la classe (Pclass)

Objectif :
Créer un graphique qui permet de comparer ces deux dimensions.

Questions :
1. Quel groupe semble avoir le meilleur taux de survie ?
2. Quelle visualisation est la plus adaptée ?

```python
survival = df.groupby(["Sex", "Pclass"])["Survived"].mean().unstack()

survival.plot(kind="bar")

plt.title("Survie selon sexe et classe")
plt.ylabel("Taux de survie")
plt.show()
```

## Exercice Matplotlib 2 - Comparer deux distributions

Comparer la distribution des âges :
- des survivants
- des non-survivants

Objectif :
Visualiser si l’âge influence la survie.

```python
plt.figure(figsize=(8, 5))

plt.hist(df[df["Survived"] == 1]["Age"].dropna(), bins=20, alpha=0.5, label="Survivants")
plt.hist(df[df["Survived"] == 0]["Age"].dropna(), bins=20, alpha=0.5, label="Non survivants")

plt.legend()
plt.title("Distribution des âges selon survie")
plt.xlabel("Âge")

plt.show()
```

## Exercice Matplotlib 3 - Relation enrichie

Créer un graphique montrant la relation entre :
- Age (x)
- Fare (y)

Mais en distinguant :
- les survivants
- les non-survivants

Objectif :
Voir si une séparation visuelle apparaît.

```python
plt.figure(figsize=(8, 5))

plt.scatter(
    df[df["Survived"] == 1]["Age"],
    df[df["Survived"] == 1]["Fare"],
    alpha=0.5,
    label="Survivants"
)

plt.scatter(
    df[df["Survived"] == 0]["Age"],
    df[df["Survived"] == 0]["Fare"],
    alpha=0.5,
    label="Non survivants"
)

plt.legend()
plt.xlabel("Âge")
plt.ylabel("Prix du billet")
plt.title("Âge vs Fare selon survie")

plt.show()
```
